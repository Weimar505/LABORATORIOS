###What is an ANN network? Explain the elements it has.
Una Red Neuronal Artificial (ANN) es un modelo computacional inspirado en el cerebro humano, compuesto por neuronas organizadas en capas (entrada, ocultas y salida), conectadas mediante pesos y bias que se ajustan durante el entrenamiento. Las neuronas procesan datos utilizando funciones de activación (como ReLU o Sigmoid) para generar salidas. El entrenamiento de la red se realiza mediante algoritmos como backpropagation, que optimizan los pesos y minimizan errores, controlados por una tasa de aprendizaje. Estas redes son clave en el aprendizaje automático, resolviendo problemas como reconocimiento de imágenes y predicciones.
###What is a single layer Neural Network? What is multilayer NN?
Una red neuronal de una sola capa tiene solo una capa de salida conectada directamente a la capa de entrada, siendo adecuada para resolver problemas simples y lineales. En cambio, una red neuronal multicapa (MLP) incluye una o más capas ocultas entre la entrada y la salida, lo que le permite modelar relaciones no lineales y resolver problemas más complejos. La principal diferencia es que las redes multicapa tienen mayor capacidad de aprendizaje gracias a sus capas ocultas y al uso de algoritmos como backpropagation para entrenarse.
###When an ANN should be used? When is not recommended?
Las ANN son ideales para problemas complejos con grandes volúmenes de datos y relaciones no lineales, como reconocimiento de imágenes, procesamiento de lenguaje natural o predicciones financieras, especialmente cuando otras técnicas no son efectivas. Sin embargo, no se recomiendan para problemas simples, datasets pequeños (porque requieren muchos datos para entrenarse), cuando se necesita interpretabilidad (ya que son modelos de "caja negra") o si se dispone de recursos computacionales limitados, ya que su entrenamiento puede ser costoso.

```python
# Importamos las librerías necesarias
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Generamos los datos según la ecuación Y = 3X + 2
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float).reshape(-1, 1)  # Aseguramos la forma (n_samples, n_features)
ys = 3 * xs + 2  # Generamos los valores de Y automáticamente

# Creamos un modelo secuencial ANN
model = keras.Sequential()

# Definimos la capa de entrada usando `keras.Input`
model.add(layers.Input(shape=(1,)))  # Una entrada con 1 característica
model.add(layers.Dense(3, activation='relu'))  # Primera capa con 3 neuronas relu=funcion de activacion
model.add(layers.Dense(3, activation='relu'))  # Segunda capa con 3 neuronas
model.add(layers.Dense(1))  # Capa de salida con 1 neurona

# Compilamos el modelo
#Entrenar usando el optimizador Adam con una tasa de aprendizaje de 0.020.02.
#Minimizar el error promedio cuadrático (MSE) como objetivo principal.
#Monitorear el progreso del error promedio cuadrático (MSE) en cada iteración del entrenamiento.
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.02), loss="mean_squared_error", metrics=['mse'])

# Entrenamos el modelo
#500 epocas
#verbose Progreso del entrenamiento, 0: No muestra ningún detalle del entrenamiento, 1: Muestra una barra de progreso por cada época., 2: Muestra una línea de texto por cada época.
model.fit(xs, ys, epochs=500, verbose=0)

# Evaluamos el modelo
loss, mse = model.evaluate(xs, ys, verbose=0)
print(f"The Mean Squared Error (MSE) is {mse:.4f}")

# Realizamos predicciones para los valores dados
x_values = np.array([[5], [3.3]])  # Valores para X = 5 y X = 3.3
predictions = model.predict(x_values)

# Mostramos los resultados
for x, y_pred in zip(x_values.flatten(), predictions.flatten()):
    print(f"Prediction for X = {x}: Y ≈ {y_pred:.4f}")
```

###What is overfitting?
El overfiting ocurre cuando un modelo aprende demasiado bien los datos de entrenamiento, incluyendo ruido y patrones irrelevantes, lo que lo hace incapaz de generalizar correctamente para nuevos datos. Esto resulta en un bajo error en el conjunto de entrenamiento pero un alto error en el conjunto de prueba.
###How it can be produced?
1. Modelo excesivamente complejo: Uso de demasiadas capas, neuronas o parámetros.
2. Datos insuficientes: Entrenar con un conjunto de datos pequeño o no representativo.
3. Entrenamiento excesivo: Demasiadas épocas que llevan al modelo a memorizar los datos en lugar de aprender patrones generales.

###What is optimizer? Describe some of the existing optimizers
Un optimizador es un algoritmo que ajusta los parámetros del modelo (pesos y bias) para minimizar la función de pérdida. Ayuda a encontrar los valores óptimos que hacen que el modelo sea más preciso.
1. SGD (Stochastic Gradient Descent): Actualiza los pesos gradualmente utilizando pequeños lotes de datos. Es simple pero puede ser lento en converger.
2. Adam (Adaptive Moment Estimation): Combina las ventajas de SGD y RMSprop.Se adapta dinámicamente a la tasa de aprendizaje, siendo rápido y eficiente.
3. RMSprop: Optimiza la tasa de aprendizaje en función de la magnitud de los gradientes recientes. Funciona bien con redes profundas y problemas con datos ruidosos.

###What is loss function? Describe some of the existing loss functions.
La función de pérdida mide la diferencia entre las predicciones del modelo y los valores reales. Es el objetivo que el modelo intenta minimizar durante el entrenamiento.
1. Mean Squared Error (MSE): Calcula el promedio del cuadrado de los errores entre las predicciones y los valores reales. Ideal para problemas de regresión.
2. Mean Absolute Error (MAE): Mide el promedio de las diferencias absolutas entre predicciones y valores reales. Menos sensible a valores atípicos que MSE.

```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder

# Montar Google Drive (solo en Google Colab)
from google.colab import drive
drive.mount('/content/drive')

```

```python
# Ruta del archivo dentro de Google Drive
file_path = '/content/drive/MyDrive/SISTEMAS EMBEBIDOS II/spam text data.csv'  # Cambia esta ruta si el archivo está en otra ubicación

# Cargamos los datos desde el archivo
data = pd.read_csv(file_path, encoding='latin-1') #Si el archivo tiene caracteres como á, é, ñ, latin-1 asegura que se lean correctamente.

data = data[['Category', 'Message']]  # Seleccionamos las columnas relevantes

# Convertimos las etiquetas de texto ('ham' o 'spam') en valores numéricos
label_encoder = LabelEncoder()
data['Category'] = label_encoder.fit_transform(data['Category'])  # 0: Ham (No Spam), 1: Spam

# Dividimos los datos en texto (X) y etiquetas (y)
X = data['Message']
y = data['Category']

# transforma el texto en un formato numérico (matriz) utilizando CountVectorizer,
vectorizer = CountVectorizer(stop_words='english', max_features=1000)  # Limita a 1000 palabras más comunes para reducir la cantidad de datos, stop_words elimina is,the, on
X_vectorized = vectorizer.fit_transform(X).toarray()
#Ajusta (fit): Aprende el vocabulario del texto, identificando todas las palabras únicas en el conjunto de datos X.
#Transforma (transform): Convierte cada texto en una matriz dispersa (sparse matrix), donde:

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42) #En este caso, 20% de los datos se usarán para prueba y 80% para entrenamiento.
#42(VALOR SEMILLA) Esto asegura que los conjuntos de entrenamiento y prueba serán los mismos cada vez que se ejecute el código, permitiendo reproducir los resultados.


# Creamos el modelo ANN para clasificar texto
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),  # Capa de entrada con 128 neuronas
    layers.Dense(64, activation='relu'),  # Capa oculta con 64 neuronas
    layers.Dense(1, activation='sigmoid')  # Capa de salida con activación sigmoide para clasificación binaria
])

# Compilamos el modelo
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamos el modelo
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Evaluamos el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {accuracy:.4f}")

# Predicción de un texto nuevo
new_texts = ["Congratulations! You've won a free ticket!", "Hi, can we reschedule our meeting?"]
new_texts_vectorized = vectorizer.transform(new_texts).toarray()
predictions = model.predict(new_texts_vectorized)
for text, pred in zip(new_texts, predictions.flatten()):
    print(f"Text: {text} -> {'Spam' if pred > 0.5 else 'Not Spam'}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
clients = pd.DataFrame(
    {
        "saldo": [
            750,
            1245,
            230,
            533,
            490,
            1000,
            190,
            900,
            600,
            50,
            1100,
            930,
            450,
            330,
            750,
        ],
        "transacciones": [
            3,
            1,
            4,
            3,
            2,
            1,
            0,
            3,
            2,
            1,
            0,
            4,
            3,
            2,
            0,
        ],
    }
)
```

```python
scaler = MinMaxScaler().fit(clients.values)
clients = pd.DataFrame(
    scaler.transform(clients.values), columns=["saldo", "transacciones"]
)
print(clients)
```

```python
kmeans = KMeans(n_clusters=2).fit(clients.values)
print(kmeans.labels_)
```

```python
clients["cluster"] = kmeans.labels_
print(kmeans.cluster_centers_, kmeans.inertia_)
```

```python
plt.figure(figsize=(6, 5), dpi=100)
colors = ["red", "orange", "blue", "black", "purple", "pink", "brown"]
print(clients)
for cluster in range(kmeans.n_clusters):
    plt.scatter(
        clients[clients["cluster"] == cluster]["saldo"],  # Pandas dataframe filter
        clients[clients["cluster"] == cluster][
            "transacciones"
        ],  # Pandas dataframe filter
        marker="x",
        s=180,
        color=colors[cluster],
        alpha=0.5,
    )
    plt.scatter(
        kmeans.cluster_centers_[cluster][0],
        kmeans.cluster_centers_[cluster][1],
        marker="o",
        s=280,
        color=colors[cluster],
    )
plt.title("Clients")
plt.xlabel("Client earnings")
plt.ylabel("Client transactions")
plt.text(1.1, 0.5, "K={}".format(kmeans.n_clusters))
plt.text(1.1, 0.3, "Inertia={}".format(round(kmeans.inertia_,2)))
plt.show()
```

###What is clustering? What is classification?
La agrupación (clustering) es una técnica de aprendizaje no supervisado que organiza datos en grupos o clusters basándose en similitudes, sin utilizar etiquetas predefinidas, como identificar clientes con comportamientos similares; ejemplos incluyen K-means y DBSCAN. Por otro lado, la clasificación es una técnica de aprendizaje supervisado que asigna datos a clases predefinidas utilizando un modelo entrenado con etiquetas, como predecir si un email es spam o no; ejemplos incluyen la regresión logística y redes neuronales. La principal diferencia es que la agrupación explora patrones ocultos en los datos, mientras que la clasificación predice categorías específicas.
#When do we use clustering and classification?
Utilizamos el agrupamiento cuando no tenemos etiquetas y queremos explorar o identificar patrones ocultos en los datos, como segmentar clientes según comportamiento. La clasificación se usa cuando tenemos etiquetas y necesitamos predecir a qué categoría pertenece un nuevo dato, como clasificar correos en spam o no spam.
#How do we use clustering with K-means?
El clustering con K-means implica dividir los datos en kk grupos asignando cada punto de datos al centroide más cercano, recalculando los centroides como el promedio de los puntos en un cluster y repitiendo este proceso hasta la convergencia. Se utiliza para agrupar datos similares, como la segmentación de clientes, basándose en características numéricas.
#Which is the difference between K-means and K-means ++?
La diferencia entre K-means y K-means++ radica en la inicialización de los centroides: mientras K-means los elige aleatoriamente, K-means++ selecciona los centroides iniciales estratégicamente, maximizando la distancia entre ellos. Esto mejora la calidad de los clusters, evita soluciones subóptimas y acelera la convergencia del algoritmo.

```python
# Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

# Datos proporcionados
data = {
    "Average_purchase_cost": [
        750, 1245, 230, 533, 490, 1000, 190, 900, 600, 50,
        1100, 930, 450, 330, 750
    ],
    "Average_purchases_credit_card": [
        3, 1, 4, 3, 2, 1, 0, 3, 2, 1,
        0, 4, 3, 2, 0
    ]
}

# Crear un DataFrame con los datos
vet_shop_data = pd.DataFrame(data)

# Escalamos los datos al rango [0, 1]
scaler = MinMaxScaler().fit(vet_shop_data.values)
vet_shop_scaled = pd.DataFrame(
    scaler.transform(vet_shop_data.values),
    columns=["Average_purchase_cost", "Average_purchases_credit_card"]
)

# Visualizamos los datos escalados
print("Datos escalados:")
print(vet_shop_scaled)

# Aplicamos K-means con 4 clusters
kmeans = KMeans(n_clusters=5, random_state=42).fit(vet_shop_scaled.values)
vet_shop_scaled["cluster"] = kmeans.labels_

# Mostramos los clusters generados
print("\nClusters asignados:")
print(vet_shop_scaled)

# Mostramos los centroides y la inercia
print("\nCentroides:")
print(kmeans.cluster_centers_)
print("\nInercia:")
print(round(kmeans.inertia_, 2))

# Graficamos los clusters
plt.figure(figsize=(8, 6))
colors = ["red", "blue", "green", "purple", "black"]

for cluster in range(kmeans.n_clusters):
    plt.scatter(
        vet_shop_scaled[vet_shop_scaled["cluster"] == cluster]["Average_purchase_cost"],
        vet_shop_scaled[vet_shop_scaled["cluster"] == cluster]["Average_purchases_credit_card"],
        marker="x",
        s=180,
        color=colors[cluster],
        alpha=0.5,
    )
    plt.scatter(
        kmeans.cluster_centers_[cluster][0],
        kmeans.cluster_centers_[cluster][1],
        marker="o",
        s=280,
        color=colors[cluster],
    )

# Personalización del gráfico
plt.title("Vet Shop Clusters")
plt.xlabel("Average Purchase Cost (scaled)")
plt.ylabel("Average Purchases with Credit Card (scaled)")
plt.text(1.1, 0.5, "K={}".format(kmeans.n_clusters))
plt.text(1.1, 0.3, "Inertia={}".format(round(kmeans.inertia_, 2)))
plt.show()

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Datos de ejemplo
datos = {
    "Costo promedio de compra (bs)": [750, 1245, 230, 533, 490, 1000, 190, 900, 600, 50, 1100, 930, 450, 330, 750],
    "Promedio de compras con tarjeta de crédito": [3, 1, 4, 3, 2, 1, 0, 3, 2, 1, 0, 4, 3, 2, 0]
}

# Crear DataFrame
df = pd.DataFrame(datos)

# Escalar características usando StandardScaler
scaler = StandardScaler()
df_escalado = scaler.fit_transform(df)

# Método del codo para determinar el número óptimo de clústeres
inercia = []
valores_k = range(1, 10)  # Probar desde 1 hasta 9 clústeres

for k in valores_k:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_escalado)
    inercia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 6))
plt.plot(valores_k, inercia, marker='o')
plt.xlabel("Número de Clústeres (k)")
plt.ylabel("Inercia")
plt.title("Método del Codo para k Óptimo")
plt.show()

# Elegir el número óptimo de clústeres (ejemplo: k=3) y ajustar el modelo final
k_optimo = 3
kmeans_final = KMeans(n_clusters=k_optimo, random_state=0)
df['Clúster'] = kmeans_final.fit_predict(df_escalado)

# Mostrar resultados
print("Datos con clústeres asignados:")
print(df[['Costo promedio de compra (bs)', 'Promedio de compras con tarjeta de crédito', 'Clúster']])
```
